# Dataset_1: Cars

```python
def cars(df):
  df=df.drop(['url', 'region_url','image_url', 'county', 'VIN', 'description', 'size'], axis=1)
  df=df.dropna(subset=['manufacturer','fuel'])
  return df
```

```python
cars_df=cars(vehicles)
```
